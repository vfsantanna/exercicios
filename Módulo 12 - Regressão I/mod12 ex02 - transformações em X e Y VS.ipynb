### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

import statsmodels.formula.api as smf

from sklearn.metrics import r2_score

%matplotlib inline
````

````python
tips = load_dataset(name='tips')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']

tips
````

````python
reg = smf.ols(formula='tip_pct ~ net_bill', data=tips).fit()

print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)


plt.axhline(y=0, color='r', linestyle='--')


plt.show()
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)

plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')

plt.show()
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
reg = smf.ols('tip_pct ~ np.log(net_bill)', data=tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')

plt.show()
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
reg = smf.ols('tip_pct ~ net_bill + np.power(net_bill, 2)', data=tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')

plt.show()
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
reg = smf.ols('np.log(tip_pct) ~ net_bill', data=tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
tips['pred_tip_pct'] = np.exp(reg.fittedvalues)

print('R-squared:', 
      tips[['pred_tip_pct', 'tip_pct']].corr().iloc[0,1]**2
     )
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], tips['pred_tip_pct'], 'r,--')

plt.show()
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
reg = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data=tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
tips['pred_tip_pct'] = np.exp(reg.fittedvalues)

print('R-squared:',
      tips[['pred_tip_pct', 'tip_pct']].corr().iloc[0,1]**2
     )
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], tips['pred_tip_pct'], 'r,--')

plt.show()
````

###  6. qual modelo resultou no melhor ajuste?

O modelo final, que utilizou o logaritmo de tip_pct como variável dependente em relação ao logaritmo de net_bill, alcançou um coeficiente de determinação, R2, de 31%. Este valor indica que aproximadamente 31% da variabilidade na porcentagem da gorjeta pode ser explicada pela variação no valor líquido da conta, quando ambos são transformados logaritmicamente. Esse resultado sugere uma relação moderada entre as duas variáveis, evidenciando que outros fatores podem também influenciar a porcentagem da gorjeta, além do valor líquido da conta.
